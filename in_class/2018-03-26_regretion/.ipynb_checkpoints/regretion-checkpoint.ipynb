```python
n = 1000
x_essays = np.random.uniform(low=0, high=1.0, size = n)
x_grades = np.random.uniform(low=0, high=1.0, size = n)

w1, w2, b = 1, 1, np.random.normal(loc=-0.7, scale=0.1, size=n)
y_accept = (w1 * x_essays + w2 * x_grades + b) > 0

plt.scatter(x_essays, x_grades, c=y_accept);
```

```python
n = 1000
x_essays = np.random.uniform(low=0, high=1.0, size = n)
x_grades = np.random.uniform(low=0, high=1.0, size = n)

w1, w2, b = 1, 1, np.random.normal(loc=-0.7, scale=0.1, size=n)
y_accept = (w1 * x_essays + w2 * x_grades + b) > 0

plt.scatter(x_essays, x_grades, c=y_accept)

x_essay_albert = 0.45
x_grade_albert = 0.45

plt.plot(x_essay_albert, x_grade_albert, 'o', markersize=20, color='purple')
```

```python
x1, x2, y = 4, 5, 0
w1, w2, b = 3, 4, -10

y_hat = ((w1 * x1) + (w2 * x2) + b) / w2
prediction = 1 if y_hat >= 0 else 0

print(y_hat, prediction, prediction == y)

color = 'blue' if prediction == y else 'red'
plt.plot(x1, x2, 'o', color=color, markersize=10)

x_values = np.linspace(0, 10, 2)
y_values = (-w1 * x_values - b)

plt.plot(x_values, y_values, color='fuchsia', linewidth=3);
```

```python
x1, x2, y = 4, 5, 0
w1, w2, b = 3, 4, -10

w1 = w1 - x1
w2 = w2 - x2
b = b - 1

y_hat = ((w1 * x1) + (w2 * x2) + b) / w2
prediction = 1 if y_hat >= 0 else 0

print(y_hat, prediction, prediction == y)

color = 'blue' if prediction == y else 'red'
plt.plot(x1, x2, 'o', color=color, markersize=10)

x_values = np.linspace(0, 10, 2)
y_values = (-w1 * x_values - b)

plt.plot(x_values, y_values, color='fuchsia', linewidth=3);
```

```python
fig_size = 16/1.5, 9/1.5
fig, ax = plt.subplots(figsize=fig_size)

x1, x2 = 4, 5
ax.plot(x1, x2, 'go', markersize=20)
legend_entries = ['Point']

x1_line = [0,10]

def plot_line(w1, w2, b, spec, name):
    x2_line = [(-w1 * x - b) / w2 for x in x1_line]
    ax.plot(x1_line, x2_line, spec, linewidth=2)
    print((w1*x1 + w2*x2 + b) / w2 >= 0)
    legend_entries.append(name)
    
w1, w2, b = 3, 4, -10
plot_line(w1, w2, b, 'k', 'Original')

w1, w2, b = 3-x1, 4-x2, -10-1
plot_line(w1, w2, b, 'r', 'Drastic')

alpha = 0.523
w1, w2, b = 3-x1*alpha, 4-x2*alpha, -10-1*alpha
plot_line(w1, w2, b, 'g', 'Slight')


w1, w2, b = 3-x1*alpha, 4, -10
plot_line(w1, w2, b, 'm--', 'w1 change')


w1, w2, b = 3, 4-x2*alpha, -10
plot_line(w1, w2, b, 'y--', 'w2 change')

w1, w2, b = 3, 4, -10-1*alpha
plot_line(w1, w2, b, 'b--', 'b change')


ax.legend(legend_entries);
```
